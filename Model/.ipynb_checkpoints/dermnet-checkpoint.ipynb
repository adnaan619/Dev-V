# 1.Importing Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten,BatchNormalization
from tensorflow.keras.layers import Conv2D,MaxPooling2D
import os
from tensorflow.keras.preprocessing import image_dataset_from_directory
```

# 2.Loading data

```python
Size = 128 #image size(128*128)
train_dataset =image_dataset_from_directory(
    "dermnet/train",#directory of the  train data
    shuffle = True,#shuffling the data in folders
    image_size = (Size ,Size ),# resize images after they are read from folder
    batch_size = 64 #Size of the batches of data to load at a time
)
test_dataset = image_dataset_from_directory(
    "dermnet/test",#directory of the  test data
    shuffle = True,
    image_size = (Size ,Size),
    batch_size = 64  
)
```

```python
classes = train_dataset.class_names
classes# All the classes in the dataset.alphanumerical order is used 
```

# 3.Data Visualization/Plotting

```python
 plt.figure(figsize = (20,20))
 for images, labels in train_dataset.take(1):
   for i in range(12):
     ax = plt.subplot(3,4,i+1)
     plt.imshow(images[i].numpy().astype("uint8"))
     plt.title(classes[labels[i]],fontsize = 10)
     plt.axis("off")
```
